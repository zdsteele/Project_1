```python
import pandas as pd
import glob
import matplotlib.pyplot as plt

path = r'D:\Data_Project\Drug_Data_Years_Zach' # use your path
all_files = glob.glob(path + "/*.csv")


df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True, sort=True)

concatenated_df.head()
```

```python
# Locking onto Colorado
geography_df = concatenated_df.set_index(['geography'])
Colorado_loc = geography_df.loc[["Colorado"], ['age_group', 'outcome', 'estimate', 'year_pair']]
```

```python
# Colorado All Ages First Time Marijuana use in the Past Year 
Colorado_First_Use_df = Colorado_loc.loc[Colorado_loc["outcome"] == "Average Annual Rate of First Use of Marijuana"]
Colorado_First_Use_df.head()
```

```python
# Colorado Ages 12 or Older Marijuana Use in the Past Month
Colorado_Mar_df = Colorado_loc.loc[Colorado_loc["outcome"] == "Marijuana Use in the Past Month"]
Colorado_age12 = Colorado_Mar_df.loc[Colorado_Mar_df["age_group"] == "12 or Older"]
Colorado_age12.drop_duplicates(subset='year_pair', keep="last").head()
```
