```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly

age_groups = ("12 or Older", "12 to 17", "18 or Older", "18 to 25", "26 or Older")
```

```python
# Graph for first time Using Marijuana in Colorado Based by Age.
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
#Read data into data frames
ftmu_file = "./First_Time_Marijuana_Use_All_Ages_Colorado_2002-2017.csv"
Colorado_First_Use_df =  pd.read_csv(ftmu_file)

Colorado_First_Use_df["year_pair"] = Colorado_First_Use_df["year_pair"].apply(lambda x: x.split('-')[0])

#Break out age ranges into separate data frames
ftmu_12_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[0]]
ftmu_12_17_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[1]]
ftmu_18_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[2]]
ftmu_18_25_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[3]]
ftmu_26_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[4]]

#Set plot size
plt.rcParams["figure.figsize"] = [20,10]

# Lambda
#ftmu_12_older_df["year_pair"] = ftmu_12_older_df["year_pair"].apply(lambda x: x.split('-')[0])
#ftmu_12_17_df["year_pair"] = ftmu_12_17_df["year_pair"].apply(lambda x: x.split('-')[0])
#ftmu_18_older_df["year_pair"] = ftmu_18_older_df["year_pair"].apply(lambda x: x.split('-')[0])
#ftmu_18_25_df["year_pair"] = ftmu_18_25_df["year_pair"].apply(lambda x: x.split('-')[0])
#ftmu_26_older_df["year_pair"] = ftmu_26_older_df["year_pair"].apply(lambda x: x.split('-')[0])

#Line Plots
plt.plot(ftmu_12_older_df["year_pair"], ftmu_12_older_df["estimate"]*100, 'ro-', linewidth=3, alpha=0.5, label=age_groups[0])
plt.plot(ftmu_12_17_df["year_pair"], ftmu_12_17_df["estimate"]*100,'bo-', alpha=0.5, linewidth=3, label=age_groups[1])
plt.plot(ftmu_18_older_df["year_pair"], ftmu_18_older_df["estimate"]*100, 'go-', alpha=0.5, linewidth=3, label=age_groups[2])
plt.plot(ftmu_18_25_df["year_pair"], ftmu_18_25_df["estimate"]*100, 'co-', alpha=0.5, linewidth=3, label=age_groups[3])
plt.plot(ftmu_26_older_df["year_pair"], ftmu_26_older_df["estimate"]*100, 'mo-', alpha=0.5, linewidth=3, label=age_groups[4])

# Add title, axis names, and legend
plt.title('First Time Marijuana Use in Colorado by Age', fontsize = "30")
plt.xlabel('Year', fontsize="30")
plt.ylabel('%', fontsize="30")
leg = plt.legend(title="Region", fontsize=20, loc="best")
leg.set_title("Region", prop = {'size': 25})


plt.axvline(x="2012", linewidth=3, color='r')
plt.grid()
plt.savefig('../Final Graphs/First Time CO Marijuana Use by Age.png')
plt.show()
```
