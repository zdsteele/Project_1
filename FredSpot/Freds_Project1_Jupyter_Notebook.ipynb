```python
#I'm not sure why I have to run it 3 times to get the final result

# Graph for first time Using Marijuana in Colorado Based by Age.
import pandas as pd
import glob
import matplotlib.pyplot as plt

path = r'C:\Users\Fred\Desktop\Project_1\FredSpot' # use your path
ftmu_file = glob.glob(path + "/First_Time_Marijuana_Use_All_Ages_Colorado_2002-2017.csv")
Colorado_First_Use_df =  pd.read_csv(ftmu_file[0])

ftmu_12_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']=="12 or Older"]
ftmu_12_17_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']=="12 to 17"]
ftmu_18_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']=="18 or Older"]
ftmu_18_25_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']=="18 to 25"]
ftmu_26_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']=="26 or Older"]

plt.plot(ftmu_12_older_df["year_pair"], ftmu_12_older_df["estimate"], color='r', alpha=0.5, marker='o', label="12 or Older")
plt.plot(ftmu_12_17_df["year_pair"], ftmu_12_17_df["estimate"], color='b', alpha=0.5, marker='o', label="12 to 17")
plt.plot(ftmu_18_older_df["year_pair"], ftmu_18_older_df["estimate"], color='g', alpha=0.5, marker='o', label="18 or Older")
plt.plot(ftmu_18_25_df["year_pair"], ftmu_18_25_df["estimate"], color='c', alpha=0.5, marker='o', label="18 to 25")
plt.plot(ftmu_26_older_df["year_pair"], ftmu_26_older_df["estimate"], color='m', alpha=0.5, marker='o', label="26 or Older")

plt.rcParams["figure.figsize"] = [20,10]

lgnd = plt.legend(title="Age Groups", fontsize="small", loc="best")

#plt.show()
```
