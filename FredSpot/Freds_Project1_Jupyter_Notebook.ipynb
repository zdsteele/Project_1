```python
### I'm not sure why I have to run it twice to get the final result ###

# Graph for first time Using Marijuana in Colorado Based by Age.

import pandas as pd
import matplotlib.pyplot as plt

age_groups = ("12 or Older", "12 to 17", "18 or Older", "18 to 25", "26 or Older")

#Read data into a data frame
ftmu_file = "./First_Time_Marijuana_Use_All_Ages_Colorado_2002-2017.csv"
Colorado_First_Use_df =  pd.read_csv(ftmu_file)

#Break out age ranges into separate data frames
ftmu_12_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[0]]
ftmu_12_17_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[1]]
ftmu_18_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[2]]
ftmu_18_25_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[3]]
ftmu_26_older_df = Colorado_First_Use_df.loc[Colorado_First_Use_df['age_group']==age_groups[4]]

#Set plot size
plt.rcParams["figure.figsize"] = [20,10]

#Line Plots
plt.plot(ftmu_12_older_df["year_pair"], ftmu_12_older_df["estimate"], 'ro-', alpha=0.5, label=age_groups[0])
plt.plot(ftmu_12_17_df["year_pair"], ftmu_12_17_df["estimate"],'bo-', alpha=0.5, label=age_groups[1])
plt.plot(ftmu_18_older_df["year_pair"], ftmu_18_older_df["estimate"], 'go-', alpha=0.5, label=age_groups[2])
plt.plot(ftmu_18_25_df["year_pair"], ftmu_18_25_df["estimate"], 'co-', alpha=0.5, label=age_groups[3])
plt.plot(ftmu_26_older_df["year_pair"], ftmu_26_older_df["estimate"], 'mo-', alpha=0.5, label=age_groups[4])

# Add title, axis names, and legend
plt.title('First Time Marijuana Use in Colorado by Age')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(title="Age Groups", fontsize="large", loc="best")

plt.show()
```
