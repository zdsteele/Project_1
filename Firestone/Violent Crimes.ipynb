```python
marijuana='../Drug_Data_Years_Zach/Marijuana_PastMonth_All_Ages_Colorado_2002-2017.csv'

marijuanaUseDF=pd.read_csv(marijuana)
marijuanaUseDF.head()
```

```python
marijuanaByYear=marijuanaUseDF.groupby('year_pair').sum()
marijuanaByYear
```

```python
marijuanaByYear['estimate']=marijuanaByYear['estimate']*1000
```

```python
marijuanaByYear
```

```python
marijuanaByYear.reset_index(inplace = True)
```

```python
marijuanaByYear['year_pair'] = marijuanaByYear['year_pair'].apply(lambda x: x.split('-')[0])
```

```python
marijuanaByYear
```

```python
marijuanaByYear=marijuanaByYear.drop('ci_lower',axis=1)
marijuanaByYear=marijuanaByYear.drop('ci_upper',axis=1)

marijuanaByYear=marijuanaByYear.drop([0,1,2,3,4,5])

marijuanaByYear
```

```python
marijuanaByYear=marijuanaByYear.rename(columns={'year_pair':'Year','estimate':'Estimate'})
marijuanaByYear
```

```python
#Crime Rate per 100,000
crimeperrate=combined_crimes[['Offense Type','2008','2009','2010','2011','2012','2013','2014','2015',
                              '2016','2017']]
crimeperrate
```

```python
crimeperrate=crimeperrate.T
crimeperrate
```

```python
crimeperrate=crimeperrate.reset_index()
crimeperrate
```

```python
crimeperrate=crimeperrate.rename(columns={'index':'Year',0:'Violent Crime',1:'Murder',
                                         2:'Aggravated Assault',3:'Non-Consensual Sex Offenses',4:'Robbery'})
crimeperrate=crimeperrate.drop([0,10])
#crimeperrate=crimeperrate.drop(crimeperrate.index[10])
crimeperrate
```

```python
marijuanaByYear
```

```python
plt.rcParams["figure.figsize"] = [20,10]

plt.plot(marijuanaByYear['Year'],marijuanaByYear['Estimate'],'ko-',linewidth='3',alpha=.5,label='Marijuana Use')
plt.plot(crimeperrate['Year'],crimeperrate['Violent Crime'],'bo-',linewidth='3',alpha=.5,label='Violent Crime')
plt.plot(crimeperrate['Year'],crimeperrate['Murder'],'ro-',linewidth='3',alpha=.5,label='Murder')
plt.plot(crimeperrate['Year'],crimeperrate['Aggravated Assault'],'yo-',linewidth='3',alpha=.5,
         label='Aggravated Assault')
plt.plot(crimeperrate['Year'],crimeperrate['Non-Consensual Sex Offenses'],'co-',linewidth='3',alpha=.5,
         label='Non-Consensual Sex Offenses')
plt.plot(crimeperrate['Year'],crimeperrate['Robbery'],'mo-',alpha=.5,label='Robbery')

plt.axvline(x='2012',linewidth='3',color='r')
plt.rc('xtick',labelsize=20)
plt.rc('ytick',labelsize=20)

plt.title('Marijuana Use and Violent Crime Rates (per 100,000)',fontsize='30')
plt.xlabel('Year',fontsize='30')
plt.ylabel('Rates (per 100,000)',fontsize='30')
plt.legend(fontsize="large", loc="best")

plt.grid()

plt.savefig('../Final Graphs/Violent Crime Rates per 100,000.png')

plt.show()
```
