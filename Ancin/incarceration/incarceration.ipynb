```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

MARIJUANA_LEGALIZATION_YEAR = 2012
```

```python
# Load in csv
crime_df = pd.read_csv("Resources/incarceration_by_state.csv")
crime_df.head(10)
```

```python
crime_df_colorado=crime_df[crime_df['jurisdiction']=='COLORADO']
crime_df_colorado.head(20)
```

```python
year = crime_df_colorado['year']
prisoner_count_colorado = crime_df_colorado['prisoner_count']
figure = plt.figure()

plt.plot(year, prisoner_count_colorado, 'r', marker='o')
plt.plot([MARIJUANA_LEGALIZATION_YEAR, MARIJUANA_LEGALIZATION_YEAR], [min(prisoner_count_colorado), max(prisoner_count_colorado)], 'k-')
plt.xlabel('YEAR')
plt.ylabel('PRISONER COUNT IN COLORADO')
plt.title('Prisoner Count in Colorado')
plt.grid()
plt.show()
figure.savefig('Incarceration_by_year_Colorado.jpg')
```

```python
crime_df_texas=crime_df[crime_df['jurisdiction']=='TEXAS']
crime_df_texas.head(20)
```

```python
year = crime_df_texas['year']
prisoner_count_texas = crime_df_texas['prisoner_count']
figure = plt.figure()

plt.plot(year, prisoner_count_texas, 'r', marker='o')
plt.plot([MARIJUANA_LEGALIZATION_YEAR, MARIJUANA_LEGALIZATION_YEAR], [min(prisoner_count_texas), max(prisoner_count_texas)], 'k-')
plt.xlabel('YEAR')
plt.ylabel('PRISONER COUNT IN TEXAS')
plt.title('Prisoner Count in Texas')
plt.grid()
plt.show()
figure.savefig('Incarceration_by_year_Texas.jpg')
```

```python
prisoner_per_capita_colorado = crime_df_colorado['prisoner_count']/crime_df_colorado['state_population']
figure = plt.figure()
plt.plot(year, prisoner_per_capita_colorado, marker='o', color='r')
plt.plot([MARIJUANA_LEGALIZATION_YEAR, MARIJUANA_LEGALIZATION_YEAR], [min(prisoner_per_capita_colorado), max(prisoner_per_capita_colorado)], 'k-')
plt.xlabel('YEAR')
plt.ylabel('PRISONER COUNT PER CAPITA IN COLORADO')
plt.title('Prisoner Count Per Capita in Colorado')

#figure.savefig('Incarceration_rate_by_year_Colorado.jpg')
plt.show()
plt.grid()
figure.savefig('Incarceration_rate_by_year_Colorado.jpg')
```

```python
prisoner_per_capita_texas = crime_df_texas['prisoner_count']/crime_df_texas['state_population']
figure = plt.figure()
plt.plot(year, prisoner_per_capita_texas, marker='o', color='r')
plt.plot([MARIJUANA_LEGALIZATION_YEAR, MARIJUANA_LEGALIZATION_YEAR], [min(prisoner_per_capita_texas), max(prisoner_per_capita_texas)], 'k-')
plt.xlabel('YEAR')
plt.ylabel('PRISONER COUNT PER CAPITA IN TEXAS')
plt.title('Prisoner Count Per Capita in Texas')

#figure.savefig('Incarceration_rate_by_year_Texas.jpg')
plt.show()
plt.grid()
figure.savefig('Incarceration_rate_by_year_Texas.jpg')
```

```python
figure = plt.figure()
plt.plot('year', 'prisoner_count', data=crime_df_colorado, label='colorado', marker='o')
plt.plot('year', 'prisoner_count', data=crime_df_texas, label='texas', marker='o')
plt.plot([MARIJUANA_LEGALIZATION_YEAR, MARIJUANA_LEGALIZATION_YEAR], 
         [min(min(prisoner_count_colorado), min(prisoner_count_texas)), 
          max(max(prisoner_count_colorado), max(prisoner_count_texas))], 'k-')
plt.xlabel('YEAR')
plt.ylabel('PRISONER COUNT')
plt.title('Prisoner Count Comparison Texas vs. Colorado')
#figure.savefig('Incarceration_by_year_texas_colorado.jpg')
plt.rcParams['figure.figsize']= [10,5]
plt.grid()
plt.show()
figure.savefig('Incarceration_by_year_texas_colorado.jpg')
```

```python
crime_df_ltd=crime_df[crime_df['jurisdiction'].isin([
    'TEXAS',
    'COLORADO', 
])]

fn = lambda r: r.prisoner_count / r.state_population
col = crime_df_ltd.apply(fn, axis=1)
crime_df_ltd = crime_df_ltd.assign(crime_per_capita=col.values) 

crime_df_ltd.head()
```

```python
figure, ax = plt.subplots(1,1)
crime_df_ltd.groupby("jurisdiction").plot(x="year", y="crime_per_capita", ax=ax, marker='o')
plt.rcParams['figure.figsize']= [10,5]
plt.legend([v[0] for v in crime_df_ltd.groupby('jurisdiction')['jurisdiction']], loc='best') 
plt.plot([MARIJUANA_LEGALIZATION_YEAR, MARIJUANA_LEGALIZATION_YEAR], 
         [min(crime_df_ltd['crime_per_capita']), max(crime_df_ltd['crime_per_capita'])],
         'k-')

plt.title('Prisoner Count Per Capita Comparison')
plt.xlabel('YEAR')
plt.ylabel('PRISONER COUNT PER CAPITA')
plt.grid()
figure.savefig('prisoner_count_per_capita_comparison.jpg')
plt.show()
```
